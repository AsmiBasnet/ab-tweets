```python
import pandas as pd
```

```python
tweets = pd.read_csv('https://raw.githubusercontent.com/AsmiBasnet/ab-tweets/refs/heads/main/Tweets.csv')
```

```python
tweets.head()
```

```python
sentiment_counts = tweets['airline_sentiment'].value_counts()
```

```python
sentiment_counts.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

sns.barplot(hue=sentiment_counts.index, y=sentiment_counts.values, palette='coolwarm')
plt.title('Airline Sentiment Count', fontsize=16)
plt.xlabel('Sentiment', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```
